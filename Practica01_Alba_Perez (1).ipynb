```python
import random
import matplotlib.pyplot as plt

# Vértices del triángulo
A = (0, 0)
B = (1, 0)
C = (0.5, 0.89)  # altura del triángulo

vertices = [A, B, C]

# Punto inicial aleatorio en el plano
x, y = random.random(), random.random()

#
xs = []
ys = []

# Iteraciones
N = 10000

for i in range(N):
    # vértice aleatoria
    vx, vy = random.choice(vertices)

    # Punto medio
    x = (x + vx) / 2
    y = (y + vy) / 2

    xs.append(x)
    ys.append(y)

# Fractal individual
plt.figure(figsize=(6,6))
plt.scatter(xs, ys, s=0.5, color="orange")
# Vertices
vx = [A[0], B[0], C[0]]
vy = [A[1], B[1], C[1]]
plt.scatter(vx, vy, color="purple", s=50)
# Etiquetas desplazadas

plt.annotate("A", xy=A, xytext=(-12, -12),
             textcoords="offset points", color="purple")

plt.annotate("B", xy=B, xytext=(12, -12),
             textcoords="offset points", color="purple")

plt.annotate("C", xy=C, xytext=(0, 12),
             textcoords="offset points", color="purple")


plt.title("Triángulo de Sierpinski")
plt.axis("equal") #Hace que las unidades sean simétricas y se cumpla que es un triangulo equilátero
plt.show()
```
